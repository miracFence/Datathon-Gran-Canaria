# CM07_1: Recursos

```python
recursos = "../../datos/CM07_1/recursos.csv"

recursos_dt = pd.read_csv(recursos)
recursos_dt.head()
```

```python
recursos_filtered = recursos_dt.fillna("No-tlfn")
recursos_filtered.head()
```

```python
for row in recursos_filtered['tlfn']:
    if row != 'No-tlfn' and row[:3] != '+34': 
        if row[:2] != '34':
            row = '+34' + row
        else:
            row = '+' + row

recursos_filtered.head(30)
```

# CM07_2: Estaciones y bicicletas

```python
bicicletas = "../../datos/CM07_2/prestamo_bicicletas.csv"

bicicletas_dt = pd.read_csv(bicicletas)
bicicletas_dt.head()
```

```python
df_agg = bicicletas_dt.groupby(['origen']).agg({'desplazamientos':sum})
g = df_agg['desplazamientos'].groupby('origen', group_keys=False)
g.nlargest(11).sort_values(ascending=False)
```
