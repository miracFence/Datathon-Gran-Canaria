# Datos inseguridad 

```python
maestro_cm35 = "../../../datos/CM35/maestro_cm35.csv"
maestro_cm35_dt = pd.read_csv(maestro_cm35)

maestro_cm35_dt.rename(columns = {'edad': 'grupo_edad'}, inplace = True)
maestro_cm35_dt
```

```python
maestro_cm35_dt['sexo'] = maestro_cm35_dt['sexo'].fillna('NON-ID')
maestro_cm35_dt['grupo_edad'] = maestro_cm35_dt['grupo_edad'].fillna('NON-ID')
maestro_cm35_dt['nacionalidad'] = maestro_cm35_dt['nacionalidad'].fillna('NON-ID')
maestro_cm35_dt['provincia'] = maestro_cm35_dt['provincia'].fillna('NON-ID')

maestro_cm35_dt
```

```python
maestro_cm35_dt['fecha'] = pd.to_datetime(maestro_cm35_dt['fecha'], format='%Y%m%d')
maestro_cm35_dt['ultima_fecha'] = pd.to_datetime(maestro_cm35_dt['ultima_fecha'], format='%Y%m%d')

maestro_cm35_dt
```

```python
maestro_cm35_dt.groupby('indicador')['valor'].mean()
```

```python
maestro_cm35_dt.groupby('indicador')['valor'].sum()
```
