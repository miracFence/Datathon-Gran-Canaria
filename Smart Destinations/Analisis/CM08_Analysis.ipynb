```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import glob
import functools
```

```python
maestro_cm08_201901_202103 = "../../../datos/CM08/maestro_cm08_201901_202103.csv"
maestro_cm08_dt = pd.read_csv(maestro_cm08_201901_202103)

maestro_cm08_dt['fecha'] = pd.to_datetime(maestro_cm08_dt['fecha'], format='%Y%m%d')

print(maestro_cm08_dt.head())

maestro_cm08_dt.describe()
```

```python
maestro_cm08_filtered = maestro_cm08_dt[maestro_cm08_dt['valor'] != 0]
print(maestro_cm08_filtered.head(10))

maestro_cm08_filtered.describe()
```

```python
print('Por tipo de visitantes: ')
print(maestro_cm08_filtered.groupby(['visitantes']).sum())

print('Por museo: ')
print(maestro_cm08_filtered.groupby(['museo']).sum())

print('Por categoría: ')
print(maestro_cm08_filtered.groupby(['categoria']).sum())

print('Por subcategoría: ')
print(maestro_cm08_filtered.groupby(['subcategoria']).sum())
```
