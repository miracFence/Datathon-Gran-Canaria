```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import glob
import functools
```

```python
maestro_cm18 = "../../datos/CM18/maestro_cm18.csv"
maestro_cm18_PIB = "../../datos/CM18/maestro_cm18_PIB.csv"

maestro_cm18_dt = pd.read_csv(maestro_cm18)
maestro_cm18_PIB_dt = pd.read_csv(maestro_cm18_PIB)

maestro_cm18_PIB_clean = maestro_cm18_PIB_dt.rename(columns = {'producto_':'producto'}, inplace = False)

combined_cm18_dt = pd.concat([maestro_cm18_dt, maestro_cm18_PIB_clean])
combined_cm18_dt
```

```python
combined_cm18_filtered = combined_cm18_dt.drop_duplicates()
combined_cm18_filtered
```

```python
combined_cm18_filtered['producto'] = combined_cm18_filtered['producto'].fillna('NON-ID')
combined_cm18_filtered['valorPrev'] = combined_cm18_filtered['valorPrev'].fillna(0)

combined_cm18_filtered
```

```python
#En lugar de diferencias, se saca una nueva columna que pueda sacar el total de diferencias entre fechas en específico 
#en un periodo en especial
combined_cm18_filtered['valorDiff'] = combined_cm18_filtered['valor'] - combined_cm18_filtered['valorPrev']

print("Suma de valores por producto: ")
print(combined_cm18_filtered.groupby('producto', as_index=False)['valor'].sum())


print("Suma de valores previos por producto: ")
print(combined_cm18_filtered.groupby('producto', as_index=False)['valorPrev'].sum())


print("Suma de diferencias de valores por producto: ")
print(combined_cm18_filtered.groupby('producto', as_index=False)['valorDiff'].sum())
```
