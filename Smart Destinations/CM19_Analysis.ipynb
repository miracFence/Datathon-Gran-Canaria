```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import glob
import functools
```

```python
maestro_cm19 = "../../datos/CM19/maestro_cm19.csv"
evol_cm19 = "../../datos/CM19/evolucion_cm19.csv"
maestro_cm19_dt = pd.read_csv(maestro_cm19)
evol_cm19_dt = pd.read_csv(evol_cm19)

evol_cm19_clean = evol_cm19_dt.rename(columns = {'anyo_': 'anyo', 'mes_': 'mes'}, inplace = False)

combined_cm19_dt = pd.concat([maestro_cm19_dt, evol_cm19_clean])
combined_cm19_dt
```

```python
combined_cm19_filtered = combined_cm19_dt.drop_duplicates()
combined_cm19_filtered
```

```python
combined_cm19_filtered['ultima_fecha'] = pd.to_datetime(combined_cm19_filtered['ultima_fecha'], format='%Y%m')
combined_cm19_filtered
```

```python
combined_cm19_filtered['lugar'] = combined_cm19_filtered['lugar'].fillna(value = 'NON-ID')
combined_cm19_filtered
```

```python
combined_cm19_filtered['valor'] = combined_cm19_filtered['valor'].fillna(value = 0)
combined_cm19_filtered['valorPrev'] = combined_cm19_filtered['valorPrev'].fillna(value = 0)
#En lugar de diferencias, se saca una nueva columna que pueda sacar el total de diferencias entre fechas en específico 
#en un periodo en especial
combined_cm19_filtered['valorDiff'] = combined_cm19_filtered['valor'] - combined_cm19_filtered['valorPrev']


print(combined_cm19_filtered.groupby('lugar', as_index=False)['valor'].sum())
print(combined_cm19_filtered.groupby('lugar', as_index=False)['valorPrev'].sum())
print(combined_cm19_filtered.groupby('lugar', as_index=False)['valorDiff'].sum())
```
